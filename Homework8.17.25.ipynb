```python
#pandas DataFrame
import pandas as pd
import matplotlib.pyplot as plt

#dataset
df = pd.read_csv('diamonds.csv', index_col=0)

#the first seven rows
print("First seven rows of the DataFrame:")
print(df.head(7))

#the last seven rows
print("\nLast seven rows of the DataFrame:")
print(df.tail(7))

#Calculate statistics for numerical columns
numerical_description = df.describe()
print("\nDescriptive statistics for numerical columns:")
print(numerical_description)

#Calculate statistics for categorical columns
categorical_columns = ['cut', 'color', 'clarity']
for column in categorical_columns:
    print(f"\nDescriptive statistics for {column}:")
    print(df[column].describe())

#unique category values for categorical columns
for column in categorical_columns:
    unique_values = df[column].unique()
    print(f"\nUnique values in {column}:")
    print(unique_values)

#Enable Matplotlib
%matplotlib inline

#histograms of each numerical data column
df.hist(figsize=(15, 10), bins=30)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

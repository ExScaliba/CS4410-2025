```python
import os
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler

os.environ["OMP_NUM_THREADS"] = "1"

# Load Iris data (150 samples, 4 features)
iris = load_iris()
X = iris.data

# Scale features (important for distance-based clustering)
X = StandardScaler().fit_transform(X)

# Compute WCSS (inertia) for k = 1..10
ks = range(1, 11)
wcss = []
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot elbow curve and save image
plt.figure(figsize=(7, 4.5))
plt.plot(list(ks), wcss, marker='o')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
# Optional: highlight k=3 (expected for Iris)
plt.axvline(3, color='red', linestyle='--', alpha=0.6, label='k = 3')
plt.legend()
plt.tight_layout()
plt.savefig(save_path, dpi=200)
plt.close()
```
